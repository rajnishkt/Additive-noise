```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha=np.sqrt(1/3)
beta=np.sqrt(2/3)
theta_pt= np.atan(beta/alpha) #radian

alpha1=np.sqrt(1/3)
beta1=-np.sqrt(2/3)
theta_pt1= 2*np.pi - np.atan(np.abs(beta1/alpha1)) #radian

alpha2=-np.sqrt(1/3)
beta2=np.sqrt(2/3)
theta_pt2=np.pi- np.atan(np.abs(beta2/alpha2)) #radian
```

```python
rad_in=[]
theta_in=[]
rad_in1=[]
theta_in1=[]
rad_in2=[]
theta_in2=[]


for i in range(10000):
    del_alpha= - np.random.normal(loc=0, scale=0.05)
    del_beta = - np.random.normal(loc=0, scale=0.05)
    if 0.999<(np.abs(alpha+del_alpha))**2+ (np.abs(beta+del_beta))**2<=1:
        r= (alpha+del_alpha)**2+ (beta+del_beta)**2
        # print(alpha+del_alpha)
        # print(beta+del_beta)
        rad_in.append(r)
        theta1= np.atan((beta+del_beta)/(alpha+del_alpha))
        theta_in.append(theta1)

for i in range(10000):
    del_alpha1= - np.random.normal(loc=0, scale=0.05)
    del_beta1 = - np.random.normal(loc=0, scale=0.05)
    if 0.999<(np.abs(alpha1+del_alpha1))**2+ (np.abs(beta1+del_beta1))**2<=1:
        r1= (alpha1+del_alpha1)**2+ (beta1+del_beta1)**2
        # print(alpha+del_alpha)
        # print(beta+del_beta)
        rad_in1.append(r1)
        theta1= np.atan((beta1+del_beta1)/(alpha1+del_alpha1))
        theta_in1.append(theta1)


for i in range(10000):
    del_alpha2= - np.random.normal(loc=0, scale=0.05)
    del_beta2 = - np.random.normal(loc=0, scale=0.05)
    if 0.999<(np.abs(alpha2+del_alpha2))**2+ (np.abs(beta2+del_beta2))**2<=1:
        r2= (alpha2+del_alpha2)**2+ (beta2+del_beta2)**2
        # print(alpha+del_alpha)
        # print(beta+del_beta)
        rad_in2.append(r2)
        theta1= np.pi- np.atan(np.abs((beta2+del_beta2)/(alpha2+del_alpha2)))
        theta_in2.append(theta1)
```

```python

plt.figure()
# Define the angle values for the circle
theta = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for the unit circle (radius = 1)
x = np.cos(theta)
y = np.sin(theta)

# Create the plot
plt.figure()
plt.plot(x, y)

# Define a specific angle for the point 
# Calculate the coordinates of the point on the circle
point_x = np.cos(theta_pt)
point_y = np.sin(theta_pt)

# Plot the point on the circle
plt.scatter(point_x, point_y, color='r', s=100, label='Ideal qubit')  # Red point 

point_x1 = np.cos(theta_pt1)
point_y1 = np.sin(theta_pt1)

# Plot the point on the circle
plt.scatter(point_x1, point_y1, color='r', s=100 )  # Red point

point_x2 = np.cos(theta_pt2)
point_y2 = np.sin(theta_pt2)

# Plot the point on the circle
plt.scatter(point_x2, point_y2, color='r', s=100)  # Red point

# points within the unit circle
# Convert polar coordinates to Cartesian coordinates
points_x = rad_in * np.cos(theta_in)
points_y = rad_in * np.sin(theta_in)

mean_point_x=np.mean(points_x)
mean_point_y=np.mean(points_y)

# Plot the random points inside the circle
plt.scatter(points_x, points_y, color='g', s=10, label='Noisy qubits')

# plt.scatter(mean_point_x, mean_point_y, color='b', s=100, label='Mean')


points_x1 = rad_in1 * np.cos(theta_in1)
points_y1 = rad_in1 * np.sin(theta_in1)

mean_point_x1=np.mean(points_x1)
mean_point_y1=np.mean(points_y1)

# Plot the random points inside the circle
plt.scatter(points_x1, points_y1, color='g', s=10)

# plt.scatter(mean_point_x1, mean_point_y1, color='b', s=100)

points_x2 = rad_in2 * np.cos(theta_in2)
points_y2 = rad_in2 * np.sin(theta_in2)

mean_point_x2=np.mean(points_x2)
mean_point_y2=np.mean(points_y2)

# Plot the random points inside the circle
plt.scatter(points_x2, points_y2, color='g', s=10)

# plt.scatter(mean_point_x2, mean_point_y2, color='b', s=100)


# Set equal scaling so the circle appears as a circle
plt.gca().set_aspect('equal', adjustable='box')

plt.grid(True)
plt.xlabel(r'$\vert 0 \rangle$')
plt.ylabel(r'$\vert 1 \rangle$')
plt.title('Scatter plot')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
